#### Расчет базовых статистик
Можно использовать метод pd.describe(), но в нем нет ряда интересующих статистик.

Напишем функцию с дополнительными стат расчетами, которая запустим по всем столбцам.

```python
def base_statistic_count(df_in):
    '''
    * Расчет базовых статистик (метрик) для переданного dataframe:
    * Результат в виде DataFrame по типу pd.desctibe()
    
    * Параметры
        df_in - pd.dataframe для которого необходимо посчитать статистики
        
    * Перечень статистик    
        Минимум
        Максимум
        Размах
        Среднеарифметическое
        Мода
        Медиана
        
        Дисперсия
        СКО
        
        Квартили Q1, Q3
        Межквартильный размах
        Выбросы
    '''
    
    stat_index_list = ['min', 
                       'max', 
                       'range', 
                       'mean',
                       'mode',
                       'median',
                       'disp',
                       'std',
                       'Q1',
                       'Q3',
                       'iqr Q3-Q1',
                       'lower_bound',
                       'upper_bound',
                       'outliers'
                      ] # изменение количества элементов изменит итоговой df_res_statisics
    columns_list = list(df_in.columns.values)
    
# Подготовка шаблона пустого DF для статистик
# Все значения при инициализации np.nan    
       
    nan_dict = {}
    for el in columns_list:
        nan_dict.update({el : np.nan})
    
    # тип данных "object", так как в ячейках будут кроме цифр, еще и списки (для выборосов)
    df_res_statisics = pd.DataFrame(nan_dict , index = stat_index_list).astype(object)

# конец инициализации

# Расчет метрик и заполнение df_res_statisics
# Пошагово берем каждый столбец и считаем для него все метрики из перечня

    for col in columns_list:    
        # сначала считаем метрики в отдельные переменные 
        # (сразу в DF не удобно - есть промежуточное использование + отладка более локальная)
        min_ = df_in[col].min()
        max_ = df_in[col].max()
        range_ = max_ - min_
        mean_ = df_in[col].mean()
        
        mode = df_in[col].mode()[0]
        median = df_in[col].median()
        
        disp = df_in[col].var()
        std_ = df_in[col].std()
        
        Q1 =  df_in[col].quantile(0.25)
        Q3 = df_in[col].quantile(0.75)
        iqr = Q3 - Q1
        
        lower_bound = Q1 - (1.5 * iqr)
        upper_bound = Q3 + (1.5 * iqr)
        outliers = df_in[col][~df_in[col].between(lower_bound, upper_bound)].to_list()

# заполнение DF результатов
        
        df_res_statisics.loc['min', col] = min_
        df_res_statisics.loc['max', col] = max_
        df_res_statisics.loc['range', col] = range_
        df_res_statisics.loc['mean', col] = mean_
        
        df_res_statisics.loc['mode', col] = mode
        df_res_statisics.loc['median', col] = median
        
        df_res_statisics.loc['disp', col] = disp
        df_res_statisics.loc['std', col] = std_

        
        df_res_statisics.loc['Q1', col] = Q1
        df_res_statisics.loc['Q3', col] = Q3
        df_res_statisics.loc['iqr Q3-Q1', col] = iqr
        
        df_res_statisics.loc['lower_bound', col] = lower_bound
        df_res_statisics.loc['upper_bound', col] = upper_bound
        df_res_statisics.loc['outliers', col] = sorted(outliers)

    return df_res_statisics
```
